```python
#Independent Variables
X = Development.iloc[:,[2,5,6,7,67]]
X = pd.DataFrame(X)

#Dependent Variable
y = Development.iloc[:,4].values
y = pd.DataFrame(y)
```

```python
#Check X and y shapes
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
#finding the sum of null values in the selected columns
X.isnull().sum()
```

```python
#Drop the Country column from X
X = X.drop(columns=["Country"])

#replacing the null values with the mean
X.fillna(X.mean(), inplace=True)
X.isnull().sum()
```

```python
#Check new shape of X
print("X shape after dropping Country:", X.shape)
X.head()
```

```python
#Train and test split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
#Check shapes of train and test data
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
#Fix null values in y
import numpy as np

y = np.array(y).ravel()
y = pd.Series(y).fillna(pd.Series(y).mean()).values

#Redo the train test split with clean y
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

print("Null values in y_train:", np.isnan(y_train).sum())
print("Null values in y_test:", np.isnan(y_test).sum())
```

```python
#Importing Linear Regression
from sklearn.linear_model import LinearRegression

#Creating the model
reg = LinearRegression()

#Fitting the model with training data
reg.fit(x_train, y_train)

print("Model trained successfully!")
```

```python
#predicting and printing the result
y_pred = reg.predict(x_test)
print(y_pred)
```

```python
#R-Squared value
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("R-Squared Score:", r2)
```

```python
#testing with few values
y_pred_few = reg.predict([[72.0, 5.2, 3341.0, 14.4]])
print(y_pred_few)
```
